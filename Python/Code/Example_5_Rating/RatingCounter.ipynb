Менее двух пропусков, +3 балла

В последней колонке контрольная: за каждый + 2.5 балла, +- 2 балла, -+ 1 балл, - 0 баллов

Остальные: + 2 балла, - вычесть 1 балл, +- 1 балл

Нужно вывести в последнем столбец итоговый рейтинг

и столбец где все оценки округлены до ближайшего целого и

все оценки превышающие 20 баллов заменены на 20 баллов


1.5 округляется как 2

```python
import math
```

```python
def calculate(line):
    rating = 0
    rating20 = 0
    
    # New solve, more beautiful
    semester = line[6:-1]
    cw = np.array(line[-1].split(','))
    Hs = np.where(np.sum(semester == 'н') <= 2, 3, 0)
    everyday_rating = 2* np.sum(semester == '+') - np.sum(semester == '-') + np.sum((semester == '+-') | (semester == '1/2'))
    cw_rating = np.sum(cw == '+') * 2.5 + np.sum(cw == '+-') * 2 + np.sum(cw == '-+')
    rating = int(np.ceil(everyday_rating + cw_rating + Hs))
    
    if rating > 20:
        rating20 = 20
    else:
        rating20 = rating
    return rating, rating20
ratings = np.apply_along_axis(calculate, 1, a).T
ratings
```

```python
df["Рейтинг"] = ratings[0]
df["Рейтинг20"] = ratings[1]
```
